```python
ds=ds.dropna(how='any',subset=['Email','CNIC','Name'])
ds=ds.reset_index(drop=True)
```

```python
cnt=0
for row in ds['POBOX']:
    if(len(str(row))!=5) or (any(c.isalpha() for c in row)):
        ds.loc[cnt,'POBOX']=np.nan
    cnt+=1
```

```python
cnt=0
for row in ds['Mobile']:
    if(len(str(row))>15) or (len(str(row))<10) or (any(c.isalpha() for c in row)):
        #print('Lenght: ',len(str(row)),'Index: ',cnt)
        ds.loc[cnt,'Mobile']=np.nan
    else:
        pass
    cnt+=1
```

```python
chars=set('@')
cnt=0
for row in ds['Email']:
    if any((x in chars) for x in row):
        pass
    else:
        ds.loc[cnt,'Email']=np.nan
    cnt+=1
        
```

re =Regular Expression
^ = must start with
$ = must end with
* = more than one
. = anything in between

```python
pattern=re.compile('^[0-9+]{5}-{0-9+}{7}-[0-9]{1}$')
count=0
for i in ds['CNIC']:
    if re.search(pattern,i):
        pass
    else:
        ds.loc[count,'CNIC']=np.nan
        
    count+=1
    
```
